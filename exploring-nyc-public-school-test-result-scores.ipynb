```python
best_math_schools = schools[(schools['average_math']/800)*100>=80].sort_values('average_math', ascending = False)
best_math_schools = best_math_schools.loc[:,['school_name', 'average_math']]
best_math_schools.head()
```

```python
schools_SAT = schools
schools_SAT['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
top_10_Schools_sorted = schools_SAT.sort_values('total_SAT', ascending = False)
top_10_schools = top_10_Schools_sorted[['school_name', 'total_SAT']].head(10)
top_10_schools
```

```python
borough = schools_SAT.groupby('borough').agg(
    num_schools = ('school_name', 'count'),
    average_SAT = ('total_SAT', 'mean'),
    std_SAT = ('total_SAT', 'std')
)
# borough = borough.head(1
borough = borough.sort_values('std_SAT', ascending = False).head(1)
borough
```
